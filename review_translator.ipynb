## Review Translator

👉 `reviews` dataset’ini load edin ve 1-yıldız rating’e sahip review’lardan bir örnek (sample) seçin.

```python
# Verileri yükle
from olist.data import Olist
data = Olist().get_data()
reviews = data["order_reviews"]
reviews
```

👀 20 adet düşük puan alan yorumdan oluşan bir örneklem seçin (rastgele) ve bunu bir listeye dönüştürün.

```python
df_new = reviews[reviews["review_score"] == 1]
df_sample = df_new.sample(n=100)
df_sample
```

🗣 Bu göreve başlamadan önce önceden yüklediğiniz `google_translator` paketini kullanarak bu yorumları çevirin:

```python
!pip install deep-translator
```

```python
from deep_translator import GoogleTranslator
import pandas as pd

translator = GoogleTranslator(source='pt', target='en')

def safe_translate(text):
    if pd.isna(text) or str(text).strip() == "":
        return None
    try:
        return translator.translate(str(text))
    except Exception as e:
        print("HATA:", e)
        return None

df_sample["translated_review"] = df_sample["review_comment_message"].apply(safe_translate)
df_sample
```

**Insights** 💡
- Bazı kötü review’lar delivery ile ilgili (`wait_time`, kaçırılan teslim tarihi, vb.)
- Ancak bazı kötü review’lar seller veya ürünle ilgili...

Peki iki olası nedeni nasıl ayırt edebiliriz?

Bu Olist’in mutlaka bilmesi gereken bir şey:
- bazı ürünleri katalogdan mı çıkarmalı?
- yoksa bazı seller’ları marketplace’ten mi kaldırmalı?


```python
def categorize(text):
    if text is None:
        return "Other"
    
    text = text.lower()

    if "delay" in text or "late" in text:
        return "Delivery Issue"
    elif "broken" in text or "damaged" in text:
        return "Damaged Product"
    elif "wrong" in text:
        return "Wrong Item"
    elif "refund" in text:
        return "Refund Issue"
    else:
        return "Other"
```

```python
df_sample["theme"] = df_sample["translated_review"].apply(categorize)
df_sample["theme"].value_counts(normalize=True) * 100
```
